```python
import pandas as pd

data=pd.read_csv("https://media.githubusercontent.com/media/kdemertzis/EKPA/main/Data/Ransomware.csv",sep='|')

data
```

```python
matrix = pps.matrix(data)
print(matrix)
```

```python
heatmap_data = matrix.pivot(index="x", columns="y", values="ppscore")
```

```python
plt.figure(figsize=(15, 13))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", vmin=0, vmax=1)
plt.title("PPS Heatmap")
plt.show()
```
